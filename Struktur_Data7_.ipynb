Membuat Binary Tree Secara Manual

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Membuat tree secara manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

# Fungsi inorder traversal
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

print("Hasil inorder traversal dari binary tree:")
inorder(root)
```

Membuat Binary Search Tree

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None
    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root

bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]
for value in data_list:
    root = bst.insert(root, value)
```

Traversal pada BST

```python
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

print("Inorder Traversal:")
inorder(root)
print("\nPreorder Traversal:")
preorder(root)
print("\nPostorder Traversal:")
postorder(root)
```

Pencarian Nilai dalam BST

```python
def search(node, key):
    if node is None or node.data == key:
        return node
    if key < node.data:
        return search(node.left, key)
    return search(node.right, key)

# Uji pencarian
key = 60
result = search(root, key)
if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")

key = 25
result = search(root, key)
if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None
    # Menyisipkan data sesuai aturan BST (kecil ke kiri, besar ke kanan)
    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root

# 7 angka unik berdasarkan identitas diri
data_list = [166, 30, 5, 77, 101, 7, 97]
bst = BinarySearchTree()
root = None
for value in data_list:
    root = bst.insert(root, value)

# Inorder: hasilkan data terurut
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

# Preorder: cetak struktur dari root ke bawah
def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

# Postorder: proses subtree sebelum root
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

print("Inorder Traversal:")
inorder(root)
print("\nPreorder Traversal:")
preorder(root)
print("\nPostorder Traversal:")
postorder(root)

# Fitur pencarian
def search(node, key):
    if node is None or node.data == key:
        return node
    if key < node.data:
        return search(node.left, key)
    return search(node.right, key)

# Cari dua digit terakhir NPM (66) - tidak ada dalam list
key = 66
result = search(root, key)
print(f"\n\nCari {key}: {'ditemukan dalam tree.' if result else 'tidak ditemukan.'}")

# Cari angka yang tidak ada dalam list
key = 69
result = search(root, key)
print(f"Cari {key}: {'ditemukan dalam tree.' if result else 'tidak ditemukan.'}")
```
